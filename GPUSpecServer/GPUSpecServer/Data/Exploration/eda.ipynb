```python
df = pd.read_csv("../Source/2025-08.csv")
df.info()
```

```python
nan_columns = df.columns[df.isnull().any()]

print("Columns with NaN values:")
print(nan_columns)
```

```python
def create_association_df(df, primary, secondary):
    options = df.groupby([primary])[secondary].unique()
    count = df.groupby([primary])[secondary].nunique()
    
    association_df = pd.DataFrame(data={f'{secondary}':options,'count':count})
    association_df = association_df[association_df['count']>1]
    return association_df
```

```python
create_association_df(df, 'architecture', 'manufacturer')
```
